```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the general Ito process
dt = 0.0001
c = 1
start_val = 1
np.random.seed(0)

# Define mu and sigma functions
def mu(S, t):
    return S + t

def sigma(S, t):
    return S * t

# Calculation of related basic parameters
n = int(1 / dt)
t = np.arange(0, n) * dt

# Calculation of the Wiener process
w = np.random.rand(n)
w = 2 * (w > 0.5) - 1  # Rescale to -1 or 1
dx = c * np.sqrt(dt)
dw = w * dx

# Calculation of the general Ito process
S = np.zeros(n)
S[0] = start_val

for i in range(1, n):
    S[i] = mu(S[i-1], t[i-1]) * dt + sigma(S[i-1], t[i-1]) * dw[i] + S[i-1]

# Plotting the result with transparent background and no grid
plt.figure(figsize=(8, 6), facecolor='none')
plt.plot(S, color='blue', linewidth=2)
plt.title("General Ito Process")
plt.xlabel("Time t")
plt.ylabel("Values of the general Ito process")
plt.xticks(ticks=np.linspace(0, n, 5), labels=np.round(np.linspace(0, 1, 5), 2))
plt.legend([f"dt = {dt}", f"c = {c}", f"s_0 = {start_val}"], loc='upper right', bbox_to_anchor=(1.45, 1))
plt.tight_layout()
plt.show()
```
