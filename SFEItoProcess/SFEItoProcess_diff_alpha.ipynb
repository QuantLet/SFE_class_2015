```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
dt = 0.0001
c = 1
start_val = 1
np.random.seed(0)

# Time setup
n = int(1 / dt)
t = np.linspace(0, 1, n)

# Drift and diffusion functions
def mu(S, t):
    return 5 * (1 - S)    #return alpha * (1-S)

def sigma(S, t):
    return 0.2 * np.sqrt(S)

# Wiener process simulation
w = np.random.choice([-1, 1], size=n)
dx = c * np.sqrt(dt)
dw = w * dx

# Simulate the Ito process
S = np.zeros(n)
S[0] = start_val

for i in range(1, n):
    S[i] = S[i-1] + mu(S[i - 1], t[i - 1]) * dt + sigma(S[i - 1], t[i - 1]) * dw[i]

# Plotting
fig, ax = plt.subplots(figsize=(10, 6), facecolor='none')
ax.plot(t, S, color='blue', linewidth=2)
ax.set_title('General Ito Process', fontsize=14)
ax.set_xlabel('Time t')
ax.set_ylabel('Values of the general Ito process')
ax.set_facecolor('none')
plt.grid(False)

# Add legend info as a text box
textstr = '\n'.join((
    f'dt = {dt}',
    f'c  = {c}',
    f'S₀ = {start_val}'
))
props = dict(boxstyle='round', facecolor='white', alpha=0.8)
ax.text(1.02, 0.95, textstr, transform=ax.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()
```
